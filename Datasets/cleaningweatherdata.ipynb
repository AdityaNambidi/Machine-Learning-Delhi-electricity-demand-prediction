```python
import csv

pf = open("./datasets/raw/Weather Data Palam.csv", 'r')
sf = open("./datasets/raw/Weather Data Sufdarjung.csv", 'r')

p = []
s = []

readerP = csv.reader(pf, delimiter=',')
readerS = csv.reader(sf, delimiter=',')
rs = [readerP, readerS]
ls = [p, s]

ignore = [2,3,5,6,8,10]

for _i in range(2):
    i = 0
    for c in rs[_i]:

        j = -1
        _r = []
        for r in c:
            j += 1
            if j in ignore:
                continue

            _r.append(r)

            

        i+=1
        ls[_i].append(_r)


print(s)


```

```python
cd = []

for r in range(len(p)):
    _r = []
    for c in range(len(p[r])):
        _r.append('')
    cd.append(_r)


for r in range(len(p)):
    for c in range(len(p[r])):
        if r == 0 :
            cd[r][c] = p[r][c]
            continue
        
        if c == 0:
            cd[r][c] = p[r][c]
            continue

        if p[r][c] == '' and s[r][c] == '':
            cd[r][c] = ''

        if p[r][c] != '' and s[r][c] == '':
            cd[r][c] = p[r][c]

        if p[r][c] == '' and s[r][c] != '':
            cd[r][c] = s[r][c]

        if p[r][c] != '' and s[r][c] != '':
            cd[r][c] = str((float(p[r][c]) + float(s[r][c]))/2)


print(cd)
        
```

```python
with open("Weather Data.csv", 'w', newline='') as f:
    wr = csv.writer(f, delimiter=',')
    wr.writerows(cd)
```
